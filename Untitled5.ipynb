```python
result = pd.read_sql_query("SELECT * FROM 门店成本", conn)
print(result)
```

```python
result = pd.read_sql_query("""
    SELECT 门店, 收入, 成本, 成本率
    FROM 门店成本
    ORDER BY 成本率 DESC
""", conn)
print(result)
```

```python
result = pd.read_sql_query("""
    SELECT 
        COUNT(*) as 门店数量,
        SUM(收入) as 总收入,
        SUM(成本) as 总成本,
        AVG(成本率) as 平均成本率,
        MAX(成本率) as 最高成本率,
        MIN(成本率) as 最低成本率
    FROM 门店成本
""", conn)
print(result)
```

```python
result = pd.read_sql_query("""
    SELECT 门店, 收入, 成本, 成本率
    FROM 门店成本
    WHERE 成本率 > (SELECT AVG(成本率) FROM 门店成本)
    ORDER BY 成本率 DESC
""", conn)
print("⚠️ 成本率高于平均的门店：")
print(result)
```

```python
result = pd.read_sql_query("""
    SELECT 
        门店,
        收入,
        成本,
        ROUND(成本率 * 100, 2) || '%' as 成本率,
        CASE 
            WHEN 成本率 < 0.72 THEN '🌟 优秀'
            WHEN 成本率 < 0.78 THEN '✅ 良好'
            ELSE '⚠️ 待改进'
        END as 评级
    FROM 门店成本
    ORDER BY 成本率
""", conn)
print(result)
```

```python
# 把评级结果导出
result = pd.read_sql_query("""
    SELECT 
        门店,
        收入,
        成本,
        ROUND(成本率 * 100, 2) || '%' as 成本率,
        CASE 
            WHEN 成本率 < 0.72 THEN '优秀'
            WHEN 成本率 < 0.78 THEN '良好'
            ELSE '待改进'
        END as 评级
    FROM 门店成本
    ORDER BY 成本率
""", conn)

result.to_excel('门店成本分析报告.xlsx', index=False)
print("✅ 分析报告已导出：门店成本分析报告.xlsx")
```

```python
import matplotlib.pyplot as plt

# 提取数据
df_chart = pd.read_sql_query("""
    SELECT 门店, 成本率 FROM 门店成本 ORDER BY 成本率
""", conn)

# 画柱状图
plt.figure(figsize=(8, 5))
plt.bar(df_chart['门店'], df_chart['成本率'])
plt.axhline(y=df_chart['成本率'].mean(), color='red', linestyle='--', label='平均成本率')
plt.ylabel('成本率')
plt.title('各门店成本率对比')
plt.legend()
plt.show()
```
